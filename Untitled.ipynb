## Here we use NLP and REGEX to eleminate the noise in Case Summary Column

```python
# NLP
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import re

for i in range(0,2319):
    example_sent =df['Case Summary'].values[i]
    stop_words = set(stopwords.words('english')) 
    word_tokens = word_tokenize(example_sent)
    #print(word_tokens)

    filtered_sentence = [w for w in word_tokens if not w in stop_words] 

    filtered_sentence = [] 

    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w) 


    #filtered_sentence.remove('.')
    #print(filtered_sentence)
    SourceList=filtered_sentence
    
# here we implement REGEX 
    OutputList = filter(lambda ThisWord: not re.match('\.|\?|\-|\;|\,|[0-9]{1,2}/[0-9]{1,2}/[0-9]{2,4}|[0-9]|/[0-9]|c/o', ThisWord),SourceList)

    ks=[]
    for ThisValue in OutputList:
        #print(ThisValue)
        ks.append(ThisValue)
    print(ks)
    #df['Case Summary'].values[i]=ks
```
